```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

# Simulacin de datos de precios de gasolina en diferentes estaciones de Durango
np.random.seed(42)  # Para reproducibilidad

# Supongamos que los precios siguen una distribucin normal aproximada con:
# Media = 25 MXN/Litro, Desviacin estndar = 1.2 MXN
precios_gasolina = np.random.normal(loc=25, scale=1.2, size=100)

# Calcular media y desviacin estndar
media = np.mean(precios_gasolina)
desviacion_estandar = np.std(precios_gasolina)

# Graficar histograma con curva normal
plt.figure(figsize=(10, 6))
count, bins, ignored = plt.hist(precios_gasolina, bins=15, density=True, alpha=0.6, color='black', edgecolor='white')

# Superponer curva normal
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, media, desviacion_estandar)
plt.plot(x, p, 'r', linewidth=2)

# Etiquetas
plt.title("Distribucin de Precios de Gasolina en Durango")
plt.xlabel("Precio por Litro (MXN)")
plt.ylabel("Densidad de Probabilidad")
plt.grid(True)
plt.show()

# Prueba de normalidad Kolmogorov-Smirnov
statistic, p_value = stats.kstest((precios_gasolina - media) / desviacion_estandar, 'norm')

# Resultados
print(f'Estadstico KS: {statistic}')
print(f'Valor p: {p_value}')
```

Este informe tiene como objetivo analizar la distribucin de los precios de la gasolina en diferentes estaciones de servicio de Durango, Mxico. Para ello, se utiliza un modelo de distribucin normal y se evala el ajuste de los datos mediante una prueba de Kolmogorov-Smirnov.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# Paso 1: Definir los datos de la tabla de contingencia
data = np.array([[120, 80, 30],
                 [150, 100, 40],
                 [100, 50, 20]])

# Crear un DataFrame para la tabla de contingencia
df = pd.DataFrame(data, columns=["Magna", "Premium", "Diesel"], index=["Centro", "Norte", "Sur"])

# Paso 2: Calcular las frecuencias esperadas
chi2, p, dof, expected = chi2_contingency(data)

# Crear un DataFrame para las frecuencias esperadas
df_expected = pd.DataFrame(expected, columns=["Magna", "Premium", "Diesel"], index=["Centro", "Norte", "Sur"])

# Paso 3: Realizar la prueba de Chi-cuadrada y mostrar el p-valor
print(f"Estadstico Chi-cuadrado: {chi2}")
print(f"P-valor: {p}")

# Paso 4: Graficar las frecuencias observadas vs esperadas con lneas
plt.figure(figsize=(10, 6))

# Graficar las frecuencias observadas
for i, row in df.iterrows():
    plt.plot(df.columns, row, marker='o', label=f'Observada - {i}', linestyle='-', markersize=8)

# Graficar las frecuencias esperadas
for i, row in df_expected.iterrows():
    plt.plot(df_expected.columns, row, marker='x', label=f'Esperada - {i}', linestyle='--', markersize=8)

# Paso 5: Aadir detalles al grfico
plt.title('Comparacin entre Frecuencias Observadas y Esperadas (Ventas de Combustible)', fontsize=14)
plt.xlabel('Tipo de Combustible', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.legend(title='Tipo de Frecuencia')
plt.grid(True)
plt.tight_layout()

# Mostrar grfico
plt.show()

# Paso 6: Interpretacin de los resultados
if p < 0.05:
    print("Existe una relacin significativa entre la zona de la ciudad y el tipo de combustible ms vendido.")
else:
    print("No existe una relacin significativa entre la zona de la ciudad y el tipo de combustible ms vendido.")
```

##Introduccin

El anlisis de la relacin entre variables categricas es fundamental para comprender los patrones de comportamiento en diversos contextos. En este estudio, se aborda la relacin entre dos variables categricas: zona de la ciudad y tipo de combustible ms vendido. Estas variables son relevantes en el contexto de la distribucin de combustibles, ya que pueden reflejar las preferencias de los consumidores en diferentes reas de la ciudad y ofrecer informacin valiosa para las estaciones de servicio y proveedores de combustible.

El objetivo de este anlisis es determinar si existe una dependencia significativa entre la zona de la ciudad (Centro, Norte, Sur) y el tipo de combustible ms vendido (Magna, Premium, Diesel). Para ello, se utiliza una prueba estadstica de Chi-cuadrada, que permite evaluar la independencia de estas variables a partir de una tabla de contingencia.

##Metodologa
Seleccin de Variables:

1:Zona de la Ciudad: Se categorizan en tres zonas: Centro, Norte y Sur de la ciudad.

2:Tipo de Combustible Ms Vendido: Las categoras de combustible incluyen Magna, Premium y Diesel.


3:Recoleccin de Datos: Se obtiene un conjunto de datos hipottico basado en las ventas de combustible de distintas zonas de la ciudad. En este caso, los datos incluyen la cantidad de combustible vendido (en litros) para cada tipo de combustible en cada zona.

La tabla de contingencia que se utilizar es la siguiente:

Zona de la ciudad	Magna	Premium	Diesel	Total

Centro	120	80	30	230

Norte	150	100	40	290

Sur	100	50	20	170

Total	370	230	90	690

4:Prueba Estadstica: Se utiliza la prueba de Chi-cuadrada para evaluar la independencia entre las dos variables categricas. Esta prueba compara las frecuencias observadas con las frecuencias esperadas bajo la hiptesis nula de que las dos variables son independientes. El estadstico de Chi-cuadrada se calcula y se obtiene el p-valor correspondiente.

##Anlisis de Resultados:

Si el p-valor es menor a 0.05, se rechaza la hiptesis nula y se concluye que hay una relacin significativa entre la zona de la ciudad y el tipo de combustible ms vendido.
Si el p-valor es mayor a 0.05, se acepta la hiptesis nula, lo que indica que no hay una relacin significativa entre las variables.
Grfico Comparativo: Para visualizar las diferencias entre las frecuencias observadas y esperadas, se genera un grfico de lneas que permite comparar las distribuciones de ventas de cada tipo de combustible en las diferentes zonas de la ciudad.

##Conclusin
El anlisis realizado utilizando la prueba de Chi-cuadrada permite evaluar si existe una relacin significativa entre la zona de la ciudad y el tipo de combustible ms vendido. Si el p-valor obtenido es menor a 0.05, se concluir que las ventas de combustibles en las diferentes zonas de la ciudad no son independientes, lo que podra indicar patrones de consumo diferenciados entre el centro, norte y sur de la ciudad.
