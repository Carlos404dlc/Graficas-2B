```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

# Simulación de datos de precios de gasolina en diferentes estaciones de Durango
np.random.seed(42)  # Para reproducibilidad

# Supongamos que los precios siguen una distribución normal aproximada con:
# Media = 25 MXN/Litro, Desviación estándar = 1.2 MXN
precios_gasolina = np.random.normal(loc=25, scale=1.2, size=100)

# Calcular media y desviación estándar
media = np.mean(precios_gasolina)
desviacion_estandar = np.std(precios_gasolina)

# Graficar histograma con curva normal
plt.figure(figsize=(10, 6))
count, bins, ignored = plt.hist(precios_gasolina, bins=15, density=True, alpha=0.6, color='black', edgecolor='white')

# Superponer curva normal
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, media, desviacion_estandar)
plt.plot(x, p, 'r', linewidth=2)

# Etiquetas
plt.title("Distribución de Precios de Gasolina en Durango")
plt.xlabel("Precio por Litro (MXN)")
plt.ylabel("Densidad de Probabilidad")
plt.grid(True)
plt.show()

# Prueba de normalidad Kolmogorov-Smirnov
statistic, p_value = stats.kstest((precios_gasolina - media) / desviacion_estandar, 'norm')

# Resultados
print(f'Estadístico KS: {statistic}')
print(f'Valor p: {p_value}')
```

##Introducción

Este informe tiene como objetivo analizar la distribución de los precios de la gasolina en diferentes estaciones de servicio de Durango, México. Para ello, se utiliza un modelo de distribución normal y se evalúa el ajuste de los datos mediante una prueba de Kolmogorov-Smirnov.

```python
# Reimportar librerías tras el reinicio del entorno
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Simulación de datos de tiempos entre llegadas de clientes (en minutos)
np.random.seed(42)

# Supongamos que el tiempo entre llegadas sigue una distribución exponencial con lambda = 1/5 (5 minutos en promedio)
tiempos_entre_llegadas = np.random.exponential(scale=5, size=100)

# Estimación de lambda (λ = 1/media)
lambda_estimado = 1 / np.mean(tiempos_entre_llegadas)

# Graficar histograma con curva exponencial
plt.figure(figsize=(10, 6))
count, bins, ignored = plt.hist(tiempos_entre_llegadas, bins=15, density=True, alpha=0.6, color='black', edgecolor='black')

# Superponer curva de densidad exponencial
x = np.linspace(0, max(tiempos_entre_llegadas), 100)
pdf_exponencial = lambda_estimado * np.exp(-lambda_estimado * x)
plt.plot(x, pdf_exponencial, 'r', linewidth=2, label='Distribución Exponencial Ajustada')

# Etiquetas
plt.title("Distribución de Tiempos entre Llegadas a una Gasolinera en Durango")
plt.xlabel("Tiempo entre Llegadas (minutos)")
plt.ylabel("Densidad de Probabilidad")
plt.legend()
plt.grid(True)
plt.show()

# Prueba de Kolmogorov-Smirnov para distribución exponencial
statistic_exp, p_value_exp = stats.kstest(tiempos_entre_llegadas, 'expon', args=(0, 1/lambda_estimado))

# Resultados
lambda_estimado, statistic_exp, p_value_exp
```

La distribución exponencial es comúnmente utilizada para modelar eventos que ocurren de manera aleatoria a lo largo del tiempo, como las llegadas de clientes en sistemas de servicio.




##Metodología

##Conclusión

En caso de que la prueba estadística indique un buen ajuste, esto sugiere que las llegadas son eventos aleatorios e independientes, lo que tiene implicaciones para la optimización de los recursos y la gestión del servicio en la estación. De lo contrario, se requerirá considerar modelos más complejos para representar la dinámica de llegadas de clientes.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# Paso 1: Definir los datos de la tabla de contingencia
data = np.array([[120, 80, 30],
                 [150, 100, 40],
                 [100, 50, 20]])

# Crear un DataFrame para la tabla de contingencia
df = pd.DataFrame(data, columns=["Magna", "Premium", "Diesel"], index=["Centro", "Norte", "Sur"])

# Paso 2: Calcular las frecuencias esperadas
chi2, p, dof, expected = chi2_contingency(data)

# Crear un DataFrame para las frecuencias esperadas
df_expected = pd.DataFrame(expected, columns=["Magna", "Premium", "Diesel"], index=["Centro", "Norte", "Sur"])

# Paso 3: Realizar la prueba de Chi-cuadrada y mostrar el p-valor
print(f"Estadístico Chi-cuadrado: {chi2}")
print(f"P-valor: {p}")

# Paso 4: Graficar las frecuencias observadas vs esperadas con líneas
plt.figure(figsize=(10, 6))

# Graficar las frecuencias observadas
for i, row in df.iterrows():
    plt.plot(df.columns, row, marker='o', label=f'Observada - {i}', linestyle='-', markersize=8)

# Graficar las frecuencias esperadas
for i, row in df_expected.iterrows():
    plt.plot(df_expected.columns, row, marker='x', label=f'Esperada - {i}', linestyle='--', markersize=8)

# Paso 5: Añadir detalles al gráfico
plt.title('Comparación entre Frecuencias Observadas y Esperadas (Ventas de Combustible)', fontsize=14)
plt.xlabel('Tipo de Combustible', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.legend(title='Tipo de Frecuencia')
plt.grid(True)
plt.tight_layout()

# Mostrar gráfico
plt.show()

# Paso 6: Interpretación de los resultados
if p < 0.05:
    print("Existe una relación significativa entre la zona de la ciudad y el tipo de combustible más vendido.")
else:
    print("No existe una relación significativa entre la zona de la ciudad y el tipo de combustible más vendido.")
```

##Introducción

El análisis de la relación entre variables categóricas es fundamental para comprender los patrones de comportamiento en diversos contextos. En este estudio, se aborda la relación entre dos variables categóricas: zona de la ciudad y tipo de combustible más vendido. Estas variables son relevantes en el contexto de la distribución de combustibles, ya que pueden reflejar las preferencias de los consumidores en diferentes áreas de la ciudad y ofrecer información valiosa para las estaciones de servicio y proveedores de combustible.

El objetivo de este análisis es determinar si existe una dependencia significativa entre la zona de la ciudad (Centro, Norte, Sur) y el tipo de combustible más vendido (Magna, Premium, Diesel). Para ello, se utiliza una prueba estadística de Chi-cuadrada, que permite evaluar la independencia de estas variables a partir de una tabla de contingencia.

##Metodología
Selección de Variables:

1:Zona de la Ciudad: Se categorizan en tres zonas: Centro, Norte y Sur de la ciudad.

2:Tipo de Combustible Más Vendido: Las categorías de combustible incluyen Magna, Premium y Diesel.


3:Recolección de Datos: Se obtiene un conjunto de datos hipotético basado en las ventas de combustible de distintas zonas de la ciudad. En este caso, los datos incluyen la cantidad de combustible vendido (en litros) para cada tipo de combustible en cada zona.

La tabla de contingencia que se utilizará es la siguiente:

Zona de la ciudad	Magna	Premium	Diesel	Total

Centro	120	80	30	230

Norte	150	100	40	290

Sur	100	50	20	170

Total	370	230	90	690

4:Prueba Estadística: Se utiliza la prueba de Chi-cuadrada para evaluar la independencia entre las dos variables categóricas. Esta prueba compara las frecuencias observadas con las frecuencias esperadas bajo la hipótesis nula de que las dos variables son independientes. El estadístico de Chi-cuadrada se calcula y se obtiene el p-valor correspondiente.

##Análisis de Resultados:

Si el p-valor es menor a 0.05, se rechaza la hipótesis nula y se concluye que hay una relación significativa entre la zona de la ciudad y el tipo de combustible más vendido.
Si el p-valor es mayor a 0.05, se acepta la hipótesis nula, lo que indica que no hay una relación significativa entre las variables.
Gráfico Comparativo: Para visualizar las diferencias entre las frecuencias observadas y esperadas, se genera un gráfico de líneas que permite comparar las distribuciones de ventas de cada tipo de combustible en las diferentes zonas de la ciudad.

##Conclusión
El análisis realizado utilizando la prueba de Chi-cuadrada permite evaluar si existe una relación significativa entre la zona de la ciudad y el tipo de combustible más vendido. Si el p-valor obtenido es menor a 0.05, se concluirá que las ventas de combustibles en las diferentes zonas de la ciudad no son independientes, lo que podría indicar patrones de consumo diferenciados entre el centro, norte y sur de la ciudad.
