```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

# Simulación de datos de precios de gasolina en diferentes estaciones de Durango
np.random.seed(42)  # Para reproducibilidad

# Supongamos que los precios siguen una distribución normal aproximada con:
# Media = 25 MXN/Litro, Desviación estándar = 1.2 MXN
precios_gasolina = np.random.normal(loc=25, scale=1.2, size=100)

# Calcular media y desviación estándar
media = np.mean(precios_gasolina)
desviacion_estandar = np.std(precios_gasolina)

# Graficar histograma con curva normal
plt.figure(figsize=(10, 6))
count, bins, ignored = plt.hist(precios_gasolina, bins=15, density=True, alpha=0.6, color='black', edgecolor='white')

# Superponer curva normal
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, media, desviacion_estandar)
plt.plot(x, p, 'r', linewidth=2)

# Etiquetas
plt.title("Distribución de Precios de Gasolina en Durango")
plt.xlabel("Precio por Litro (MXN)")
plt.ylabel("Densidad de Probabilidad")
plt.grid(True)
plt.show()

# Prueba de normalidad Kolmogorov-Smirnov
statistic, p_value = stats.kstest((precios_gasolina - media) / desviacion_estandar, 'norm')

# Resultados
print(f'Estadístico KS: {statistic}')
print(f'Valor p: {p_value}')
```

##Introducción

Este informe tiene como objetivo analizar la distribución de los precios de la gasolina en diferentes estaciones de servicio de Durango, México. Para ello, se utiliza un modelo de distribución normal y se evalúa el ajuste de los datos mediante una prueba de Kolmogorov-Smirnov.


##Pasos a seguir

3:Visualización: Se elabora un histograma con la curva de distribución normal.

4:Prueba de normalidad: Se utiliza la prueba de Kolmogorov-Smirnov para determinar el ajuste a la distribución normal.
