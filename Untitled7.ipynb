#Parte 1: Comparación de Medias (Hoja: Comparacion_Medias)

##Hipótesis:
○ H0: No hay diferencia significativa entre las calificaciones antes y después.

○ H1: Existe una diferencia significativa entre las calificaciones antes y después.

 indican que hay una diferencia significativa entre las calificaciones antes y después (rechazamos H₀ con un valor p ≈ 0).

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Datos de satisfacción por sucursal
satisfechos_s1 = 100  # Número de clientes satisfechos en Sucursal 1
total_s1 = 100        # Total de clientes en Sucursal 1

satisfechos_s2 = 84   # Número de clientes satisfechos en Sucursal 2
total_s2 = 120       # Total de clientes en Sucursal 2

# Proporciones de satisfacción
p1 = satisfechos_s1 / total_s1
p2 = satisfechos_s2 / total_s2

# Proporción combinada
p_comb = (satisfechos_s1 + satisfechos_s2) / (total_s1 + total_s2)

# Estadístico Z
z_score = (p1 - p2) / np.sqrt(p_comb * (1 - p_comb) * (1/total_s1 + 1/total_s2))

# Valor p (dos colas)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Gráfico de barras de satisfacción
labels = ['Sucursal 1', 'Sucursal 2']
values = [p1, p2]

plt.figure(figsize=(6, 4))
plt.bar(labels, values, color=['blue', 'orange'])
plt.ylabel('Proporción de satisfacción')
plt.title('Proporción de Clientes Satisfechos por Sucursal')
plt.ylim(0, 1)
plt.show()

# Resultados
z_score, p_value
```
