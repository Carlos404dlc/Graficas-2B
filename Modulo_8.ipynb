En conclusión, los resultados indican que el deporte es una actividad valorada por los encuestados debido a sus beneficios físicos y personales, pero su práctica se ve limitada por factores como la falta de tiempo y la carencia de espacios adecuados. Esto evidencia la necesidad de implementar estrategias que fomenten la actividad física, como el desarrollo de infraestructura deportiva accesible y la promoción de hábitos saludables en la población.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Datos del problema (ajustados a la encuesta)
media_muestral_dias = 3.77  # Promedio de días de práctica
media_muestral_anos = 2.43  # Promedio de años practicando
desviacion_estandar_dias = 1.5  # Aproximación basada en la variabilidad observada
desviacion_estandar_anos = 1.0  # Aproximación basada en la variabilidad observada
tamano_muestra = 30
nivel_confianza = 0.95
valor_z = 1.96  # Para un nivel de confianza del 95%

# Calcular el error estándar
def calcular_intervalo(media, desviacion):
    error_estandar = desviacion / np.sqrt(tamano_muestra)
    margen_error = valor_z * error_estandar
    limite_inferior = media - margen_error
    limite_superior = media + margen_error
    return error_estandar, margen_error, limite_inferior, limite_superior

# Cálculo de intervalos para los días de práctica
error_estandar_dias, margen_error_dias, limite_inf_dias, limite_sup_dias = calcular_intervalo(media_muestral_dias, desviacion_estandar_dias)

# Cálculo de intervalos para los años de experiencia
error_estandar_anos, margen_error_anos, limite_inf_anos, limite_sup_anos = calcular_intervalo(media_muestral_anos, desviacion_estandar_anos)

# Graficar
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Gráfico para días de práctica
x_dias = np.linspace(media_muestral_dias - 4 * desviacion_estandar_dias, media_muestral_dias + 4 * desviacion_estandar_dias, 1000)
y_dias = norm.pdf(x_dias, media_muestral_dias, error_estandar_dias)
axes[0].plot(x_dias, y_dias, label='Distribución Normal', color='blue')
axes[0].fill_between(x_dias, y_dias, where=(x_dias >= limite_inf_dias) & (x_dias <= limite_sup_dias), color='gray', alpha=0.5, label='Intervalo de Confianza (95%)')
axes[0].axvline(x=limite_inf_dias, color='red', linestyle='--', label=f'Límite Inferior: {limite_inf_dias:.2f}')
axes[0].axvline(x=limite_sup_dias, color='red', linestyle='--', label=f'Límite Superior: {limite_sup_dias:.2f}')
axes[0].axvline(x=media_muestral_dias, color='green', linestyle='-', label=f'Media Muestral: {media_muestral_dias:.2f}')
axes[0].set_title('Intervalo de Confianza del 95% - Días de Práctica')
axes[0].set_xlabel('Días de Práctica')
axes[0].set_ylabel('Densidad de Probabilidad')
axes[0].legend()
axes[0].grid(True)

# Gráfico para años de experiencia
x_anos = np.linspace(media_muestral_anos - 4 * desviacion_estandar_anos, media_muestral_anos + 4 * desviacion_estandar_anos, 1000)
y_anos = norm.pdf(x_anos, media_muestral_anos, error_estandar_anos)
axes[1].plot(x_anos, y_anos, label='Distribución Normal', color='blue')
axes[1].fill_between(x_anos, y_anos, where=(x_anos >= limite_inf_anos) & (x_anos <= limite_sup_anos), color='gray', alpha=0.5, label='Intervalo de Confianza (95%)')
axes[1].axvline(x=limite_inf_anos, color='red', linestyle='--', label=f'Límite Inferior: {limite_inf_anos:.2f}')
axes[1].axvline(x=limite_sup_anos, color='red', linestyle='--', label=f'Límite Superior: {limite_sup_anos:.2f}')
axes[1].axvline(x=media_muestral_anos, color='green', linestyle='-', label=f'Media Muestral: {media_muestral_anos:.2f}')
axes[1].set_title('Intervalo de Confianza del 95% - Años de Experiencia')
axes[1].set_xlabel('Años de Experiencia')
axes[1].set_ylabel('Densidad de Probabilidad')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```
