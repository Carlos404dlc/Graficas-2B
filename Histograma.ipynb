```python
import matplotlib.pyplot as plt
import numpy as np
# Datos simulados (calificaciones de alumnos)
data = np.random.normal(loc=80, scale=20, size=1000) # Media 70, desviación estándar 10
# Crear histograma
plt.hist(data, bins=30, color='red', alpha=0.9, edgecolor='blue')
plt.title('Histograma de Calificaciones')
plt.xlabel('Calificaciones')

1

plt.ylabel('Frecuencia')
plt.show()
```
