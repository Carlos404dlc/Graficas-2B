```python
import matplotlib.pyplot as plt
import numpy as np
# Datos simulados (calificaciones de alumnos)
data = np.random.normal(loc=80, scale=20, size=1000) # Media 70, desviación estándar 10
# Crear histograma
plt.hist(data, bins=30, color='red', alpha=0.9, edgecolor='blue')
plt.title('Histograma de Calificaciones')
plt.xlabel('Calificaciones')

1

plt.ylabel('Frecuencia')
plt.show()
```

```python
categories = ['Fisica', 'Ciencia', 'Historia', 'Geografia', 'Matematicas']
frequencies = [90, 130, 50, 10, 110]
plt.bar(categories, frequencies, color='black', alpha=0.7, edgecolor='green')
plt.title('Diagrama de Barras: Alumnos por Materia')
plt.xlabel('Materias')
plt.ylabel('Cantidad de Alumnos')
plt.show()
```
