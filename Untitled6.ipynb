```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parámetros de la distribución binomial
n = 10  # Número de intervalos (podría representar intentos de accidente en 10 min)
p = 0.05  # Probabilidad de que ocurra un accidente en cada intento

# Simulación de la distribución binomial
x = np.arange(0, n+1)
y = stats.binom.pmf(x, n, p)

# Graficar la distribución binomial
plt.bar(x, y, color='blue', alpha=0.6, label='Distribución Binomial')
plt.xlabel('Número de accidentes en 10 minutos (k)')
plt.ylabel('Probabilidad')
plt.title('Distribución Binomial de Accidentes en 10 Minutos')
plt.legend()
plt.grid()
plt.show()

# Cálculo del valor esperado y la varianza
media = n * p
varianza = n * p * (1 - p)

print(f"Valor Esperado (Media): {media}")
print(f"Varianza: {varianza}")

# Interpretación de resultados
moda = x[np.argmax(y)]  # El valor de k con mayor probabilidad
print(f"El número de accidentes más probable en 10 minutos es: {moda}")
```

**Contexto:**

dado que la probabilidad de que ocurra un accidente en cada minuto es de 0.05 (probabilidad
de éxito en cada intento).

**conclusiones:**

Distribución de Accidentes

Valor Esperado y Varianza
