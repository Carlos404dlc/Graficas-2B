```python
x = np.random.uniform(6, 20, 50) # Horas de estudio
y = 5 * x + np.random.normal(scale=5, size=50) # Calificaciones con algo de ruido
plt.scatter(x, y, color='black', alpha=0.7)
plt.title('Diagrama de Dispersi√≥n: Estudio vs Calificaciones')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.show()
```
